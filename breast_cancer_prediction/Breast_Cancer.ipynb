```python
import pandas as pd
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer-wisconsin/wdbc.data'

features = ['radius', 'texture', 'perimeter', 'area', 'smoothness', 'compactness',
            'concavity', 'concave_poinits', 'symmetry', 'fractal_dimension']
column_names = ['id', 'diagnosis']

for attr in ['mean', 'ste', 'largest']:
  for feature in features:
    column_names.append(feature + "_" + attr)
```

```python
pd.set_option('display.max_columns', None)
dataset = pd.read_csv(url, names=column_names)
dataset.sample(5)
```

```python
import missingno
missingno.matrix(dataset)
```

```python
dataset.isna().sum()
```

```python
dataset.describe()
```

```python
dataset.describe(include='O')
```

```python
dataset['diagnosis']=dataset['diagnosis'].map({'B':0,'M':1})
```

```python
for col in dataset.columns:
    print( col,':', dataset[col].nunique() )
    print(dataset[col].value_counts().nlargest(5))
    print('\n' + '*' * 20 + '\n')
```

```python
X=dataset.drop(columns=['id','diagnosis'])
y=dataset['diagnosis']
print(X.head())
print(y.head())
```

```python
from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=17)
```

# use only "radius_mean" feature

```python
from sklearn.linear_model import LogisticRegression
model_1=LogisticRegression(solver='sag')
model_1.fit(X_train[['radius_mean']], y_train)
```

```python
y_pred_1=model_1.predict(X_test[['radius_mean']])
y_pred_1
```

```python
accuracy1=model_1.score(X_test[['radius_mean']],y_test)
accuracy1
```

# use all data

```python
model_2=LogisticRegression(solver='sag')
model_2.fit(X_train,y_train)
```

```python
y_pred_2=model_2.predict(X_test)
y_pred_2
```

```python
accuracy2=model_2.score(X_test,y_test)
accuracy2
```

# use all data with standard scaler

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_train_scaled=scaler.fit_transform(X_train)
X_train_scaled=pd.DataFrame(X_train_scaled,columns=X_train.columns)

X_test_scaled=scaler.fit_transform(X_test)
X_test_scaled=pd.DataFrame(X_test_scaled,columns=X_test.columns)
```

```python
model_3=LogisticRegression(solver='sag')
model_3.fit(X_train_scaled,y_train)
```

```python
y_pred_3=model_3.predict(X_test_scaled)
y_pred_3
```

```python
accuracy3=model_3.score(X_test_scaled,y_test)
accuracy3
```

# compare between the three model

```python
[accuracy1,accuracy2,accuracy3]
```
